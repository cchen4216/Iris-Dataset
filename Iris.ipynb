```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# 导入数据集并简单查看数据内容
df = pd.read_csv('./bezdekIris.data.txt')
```

```python
df.head()
```

Iris 一共有三种 `class_name`，我们先按照 `class_name` 对 dataframe 进行分组，然后分别对 `sepal_len`, `sepal_wid`, `petal_len`, `petal_wid` 四个 feature 进行统计分析。

从统计结果来看，三种 iris 的 petal length 和 width 存在非常明显的差别。

```python
df.groupby('class_name').agg('describe').transpose()
```
